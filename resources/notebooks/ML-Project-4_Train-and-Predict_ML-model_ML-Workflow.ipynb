# Machine Learning Project: 4. Train & predict - with machine learning model

This section shows how to:
* **Train** the ML model with training data set,
* **Make predictions** using ML model with test data set.

```python
# Load the data using 'pandas'
import pandas as pd
df_house = pd.read_csv("data/house.csv")
```

```python
# Split the data into 'features' and 'labels'
X = df_house.drop("SalePrice", axis=1)
y = df_house["SalePrice"]
```

```python
# Split the data into 'training set' and 'test set'
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2)
```

```python
# Create machine learning model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Train the ML model
model.fit(X_train, y_train)
```

```python
# Make predictions using ML model
y_preds = model.predict(X_test)
```
